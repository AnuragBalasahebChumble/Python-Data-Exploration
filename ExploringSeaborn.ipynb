###Note:
- Now will see how to plot a dig for data set which consists of a measurement.
- Here we are trying to find the precession value, so the measurement is of the same dimension/object/event but it will have slight variations.
-We are going to use another default dataset made available by seaborn library

```python
fmri = sns.load_dataset('fmri')
fmri.sample(20)
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line')
```

###Note:
- In statistics there is a concept of confidence interval. Here it has been put to used. Confidence interval of 95% is used while plotting such line graphs with seaborn library.(it means 95% chance is of the value to lie in specified range)

- So for timepoint 5.0 approx 0.18-0.10 is the confidence interval.

```python
# If you don't want the confidence intervals to be present on the graph, you can use "CI" variable.
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', ci=False)
```

```python
# We can also pass Standard Deviation(variation from the mean value) as a value to ci variable.
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', ci='sd')  
```

###Conclusion:
- So this how we can plot the data which have multiple value for same time/ same parameter value.

```python
sns.relplot(x='timepoint', y='signal', estimator=None, kind='line', data=fmri)
```

###Note:
- You see if we avoid the confidence interval completely using **estimator set to None**, the curve we get is not a smooth curve, and so not very helpful either.

```python
# Note we will make use of a categorical valriable event for hue value.
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', hue='event')
```

###Note
- Red line is the projection of cue type of event
- Blue line is the projection of stim type of event

```python
# Let's bring in the style variable into play.

sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='region', style='event')
```

```python
# Now, it's right time to introduce markers. Markers can increase the readability of the graph.

sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='event', style='region', markers=True)
```

```python
# We can remove the dashes usning 'dashes= False'

sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='event', style='region', markers=True, dashes=False)
```

```python
# We can also use same parameter to be a value for 'style' & 'hue' to distinguish more clearly.

sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='event', style='event' )
```

```python
# Now we'll plot the graph with respect to subjects.
# This is just to recall that we have a paramter called subject.

fmri.sample(3)
```

```python
only_cue = fmri[fmri.event=='cue']
```

```python
# This will give error ask you to set, estimator = None
# When you set estimator = None,  you don't get a plot with "confidence Interval".

sns.relplot(x='timepoint', y='signal', kind='line', hue='region', data=only_cue, units='subject')
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', hue='region', data=only_cue, units='subject', estimator=None)
```

###Note:
- Sometimes color sceme provided by the hue variable is poor.
- We can use color palette to increase the readability of the graph.

```python
# Now, let's use size varaible with coherence.

sns.relplot(x='time', y='firing_rate', data=df.query("align=='dots'"), kind='line', size='coherence', hue='coherence', style='choice')
```

```python
# We'll cahnge the size of the line using sizes

sns.relplot(x='time', y='firing_rate', data=df.query("align=='dots'"), kind='line', size="coherence",  sizes=(1, 7), hue='coherence', style='choice')
```

###**BOXPLOT** 

- They are used to display descriptive stats.

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

###Takeaway's
- Reading from the first blue box.
1. dots represents the outliers
2. Top most 'T'-> horizontal line represents 75% quartile. i.e. 75%  of the bills are less than 30.
3. Inverted 'T' -> horizontal line represents min value for bill on thursday.
4. The blue box represents 50% of the data. The Top most 'T' gives 25% of the data & Inverted 'T' represents 25% of the data.

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box', hue='sex')
```

###Observation:
