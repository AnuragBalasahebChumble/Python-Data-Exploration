**Anonymous function within user defined functions**
- Lambda functions are best use within other higher-order functions.

```python
def new_func(x):
  return lambda y : x+y
# here we have a user define function that takes a lamda function as an input
# In this example the normal python function new_func(x) takes one arguement 'x'
# This arguement is then added to some unknown arguement which is supllied 
# the lambda function.
```

```python
t = new_func(4)
print(t(8))
# The lamda function used within the function new_func() is called whenever
# the higher order new_func() is called.
# the first thing we are doing here passing the value to variable "x"
# and then we are printing the value of x+y
```

```python
t = new_func(3)
print(t(4))
```

**Using lambda function with filter(), map(), & reduce()**
- filter(): 
1. Used to filter the given iterables(lists, sets etc) with the help of another finction passed as an arguement to test all the elements to be true or false.
2. Syntax: *filter*(function, iterables)
3. Mkae sure it return the iterable, otherwise you don't see a answer. 

```python
my_list=[1, 2, 3, 4, 5, 6, 7, 9]
new_list = list(filter(lambda a: a/3==2, my_list)) 
# it will iterate through list usng variable a, & append the value 
# to list which satisfies the lambda expression.        
```

```python
print(new_list)
```

- **Map():** Applies to a given function to all the iterables and returns a new list.
- Syntax: map(func, iterable)
- It should also return a list object.

```python
my_list=[1, 2, 3, 4, 5, 6, 7, 9]
new_list = list(map(lambda a: a/3!=2, my_list))
new_list
# It returns boolean values with respect to the expression passed.
```

- **reduce()**: Applies some other function to a list of elements that are passed as a parameter to it and finally **returns a single value.**
- Syntax: reduce(function, sequence)

```python
from functools import reduce
my_list=[1, 2, 3, 4, 5, 6, 7, 9]
reduce(lambda a, b: a+b, my_list)
# function adds values recursively and reduces the list to sum of it's element.
```

###**2nd use case**
- Applying a Function on a Pandas DataFrame

```python
data = pd.DataFrame({
    'EmployeeName': ['Callen Dunkley', 'Sarah Rayner', 'Jeanette Sloan', 'Kaycee Acosta', 'Henri Conroy', 'Emma Peralta', 'Martin Butt', 'Alex Jensen', 'Kim Howarth', 'Jane Burnett'],
    'Department': ['Accounting', 'Engineering', 'Engineering', 'HR', 'HR', 'HR', 'Data Science', 'Data Science', 'Accounting', 'Data Science'],
    'HireDate': [2010, 2018, 2012, 2014, 2014, 2018, 2020, 2018, 2020, 2012],
    'Sex': ['M', 'F', 'F', 'F', 'M', 'F', 'M', 'M', 'M', 'F'],
    'Birthdate': ['04/09/1982', '14/04/1981', '06/05/1997', '08/01/1986', '10/10/1988', '12/11/1992', '10/04/1991', '16/07/1995', '08/10/1992', '11/10/1979'],
    'Weight': [78, 80, 66, 67, 90, 57, 115, 87, 95, 57],
    'Height': [176, 160, 169, 157, 185, 164, 195, 180, 174, 165],
    'Kids': [2, 1, 0, 1, 1, 0, 2, 0, 3, 1]
})
```

```python
data.sample(3)
```

 **Create two columns for storing the employees’ first and last names separately**

```python
data['Name']= data.EmployeeName.apply(lambda a : a.strip().split(" ")[0])
data['LastName'] = data.EmployeeName.apply(lambda a : a.strip().split(" ")[1])
data.sample(3)
```

**Calculate age and create a column for it**

```python
from datetime import datetime, date

def calculate_age(birthdate):
    birthdate = datetime.strptime(birthdate, '%d/%m/%Y').date()
    today = date.today()
    return today.year - birthdate.year - (today.month < birthdate.month)
```

```python
data['Age'] = data['Birthdate'].apply(calculate_age)
data.sample(2)
```

**Provide a Body Mass Index (BMI) for every employee**
- We need to define a function that calculates the Body Mass Index (BMI)
- The formula for the BMI is weight in kilograms divided by height in meters **squared.**
- Because the employees’ heights are measured in centimeters, we need to divide the heights by 100 to obtain the heights in meters.

```python
data['HeightInMeters']= data.Height.apply(lambda a : a/100) 
```

**Note**
- It is the apply() that iterates through the Series of DataFrame.
- With Lamda function within **apply()** you iterate over **One series object.**
- But you won't be able to perform, say multiplication **operation between two series object.**
- For that you'll have to write separate function, & then pass it in lambda function.

```python
data.sample(2)
```

```python
def bmi(wt, ht):
  return np.round((wt/ht**2), 2) ##height is meter sqaure

data['BMI'] = data.apply(lambda a: bmi(a.Weight, a.HeightInMeters), axis=1)

# Here we are performing operation between two series
# The axis=1 argument means to iterate over rows in the DataFrame.
```

- Parameters

- Function to apply to each column or row.

- axis{0 or ‘index’, 1 or ‘columns’}, default 0
- Axis along which the function is applied:

- 0 or ‘index’: apply function to each column.

- 1 or ‘columns’: apply function to each row.


```python
data.sample(1)
```

```python
def indicator(bmi):
    if (bmi < 18.5):
        return 'Group One'
    elif (18.5 <= bmi < 25):
        return 'Group Two'
    elif (25 <= bmi < 30):
        return 'Group Three'
    else:
        return 'Group Four'

data['BMI_Indicator'] = data['BMI'].apply(indicator)
display(data)
```

Let's see how to convert the string series into Datetime

```python
# Just note all different format in which we are writting the date, so to undersatnd the power of 'to_datetime()'.
dates = pd.Series(['December 22nd, 2019', "2019, 12, 22", "This is date", 'Dec 23rd 2019'])
dates
```

```python
pd.to_datetime(dates, errors='coerce')
#Just see how it understands & converts the str data to datetime data.
```

**errors : {'ignore', 'raise', 'coerce'}, default 'raise'**


- If 'raise', then invalid parsing will raise an exception.


- If 'coerce', then invalid parsing will be set as NaT.(not a time)


- If 'ignore', then invalid parsing will return the input.

```python
unixtime = [12345678, 22334455, 22331112]
```

```python
pd.to_datetime(unixtime, unit='s')
```

- The unit of the arg (D,s,ms,us,ns) denote the unit, which is an
integer or float number. 

- Example, with unit='ms' and origin='unix' (the default), this
would calculate the number of milliseconds to the unix epoch start.


```python
pd.to_datetime(unixtime, unit='ms')
```
